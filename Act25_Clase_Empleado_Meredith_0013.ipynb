```python
# Definimos la clase Empleado
class Empleado:
    # El método __init__ inicializa los atributos de la clase
    def __init__(self):
        self.nombre = ""
        self.puesto = ""
        self.salario = 0.0

    # Función para pedir la información al usuario
    def leer_datos(self):
        print(f"\n--- Ingresando datos del empleado ---")
        self.nombre = input("Escribe el nombre: ")
        self.puesto = input("Escribe el puesto: ")
        # Remove commas from the input string before converting to float
        salario_str = input("Escribe el salario: ").replace(',', '')
        self.salario = float(salario_str)

    # Función para imprimir la información en pantalla
    def mostrar_datos(self):
        print("\n--- Información del Empleado ---")
        print(f"Nombre: {self.nombre}")
        print(f"Puesto: {self.puesto}")
        print(f"Salario: ${self.salario:.2f}")

# --- Bloque principal del programa ---

# 1. Creamos los 2 objetos (instancias de la clase Empleado)
empleado1 = Empleado()
empleado2 = Empleado()

# 2. Interactuamos con el primer objeto
print("Datos para el primer empleado:")
empleado1.leer_datos()

# 3. Interactuamos con el segundo objeto
print("\nDatos para el segundo empleado:")
empleado2.leer_datos()

# 4. Mostramos los resultados finales
print("\n========================================")
print("RESUMEN DE NÓMINA")
empleado1.mostrar_datos()
empleado2.mostrar_datos()
print("Programa realizado por Meredith Aguirre NL = 2")
```
